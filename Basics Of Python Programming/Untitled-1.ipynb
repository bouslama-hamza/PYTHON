```python
#exercice 11

n   =   int(input("Donnez un nombre :"))

if n<0 :

    print("n est négatif, erreur, recommencez")

else:

    print("n est positif, suite du programme")

    d = 2

    while n>1 :

        while n%d==0 :

            n = n//d

            print("facteur trouvé :", d)

        d += 1

print("Fin de programme")
```

```python
def diviseur(n):

    L = []

    for i in range(1,n+1):

        if n % i == 0 :
            
            L = L + [i]
            
    return L

a = int(input("donnez un nombre :"))

diviseur(a)
```
